```python
def plot_hist(sample_df, full_df, var, sample_mean=False, true_mean=False):
    tmp = sample_df[var].dropna()
    plt.hist(tmp, color='grey', alpha=.4)
    plt.title(var)
    if sample_mean is True: plt.axvline(np.mean(tmp), color='dodgerblue')
    if true_mean is True: plt.axvline(np.mean(full_df[var]), color='tomato')
```
